# Results Analysis

Analyze training results and model performance.

```python
import json
import numpy as np
import pandas as pd
```

```python
# Load evaluation results
with open('../results/evaluation.json', 'r') as f:
    results = json.load(f)

print('Model evaluation results:')
for metric, value in results.items():
    if isinstance(value, float):
        print(f'  {metric}: {value:.4f}')
    else:
        print(f'  {metric}: {value}')
```

```python
# Compare against baseline
baseline = {
    'accuracy': 0.921,
    'f1_macro': 0.876,
    'precision': 0.889,
    'recall': 0.864,
}

print('\nComparison with baseline:')
for metric in baseline:
    if metric in results:
        diff = results[metric] - baseline[metric]
        sign = '+' if diff > 0 else ''
        print(f'  {metric}: {results[metric]:.4f} ({sign}{diff:.4f})')
```

```python
# Per-class performance summary
if 'per_class' in results:
    df = pd.DataFrame(results['per_class']).T
    df.index.name = 'Class'
    print('\nPer-class metrics:')
    print(df.to_string())
```
